```python
import pandas as pd
import numpy as np
df["logConfirmed"] = np.log10(df[df["Confirmed"] != 0]["Confirmed"])
df["logDeaths"] = np.log10(df[df["Deaths"] != 0]["Deaths"])
df["logRecovered"] = np.log10(df[df["Recovered"] != 0]["Recovered"])
```

```python
# Here new column is added that is log value of the column Confirmed, Deaths, Recovered excluding zero.
```

```python
# cdr['Confirmed'] = cdr['Confirmed'].astype('int64')
# cdr['Deaths'] = cdr['Deaths'].astype('int64')
# cdr['Recovered'] = cdr['Recovered'].astype('int64')
# cdr
cdr = df
cdr
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns 
plt.hist(cdr["logConfirmed"],bins=10,color='blue',label="Confirmed Cases")
plt.hist(cdr["logDeaths"],bins=10,alpha =0.6, color='red',label="Death Cases")
plt.hist(cdr["logRecovered"],bins=10,alpha =0.6,color='violet',label="Recovered Cases")
plt.legend()
plt.xlabel("Log(Cases)")
plt.show()
```

```python
# Finally, ths histogram is plotted using seaborn and matplotlib.
```

```python
import numpy as np
sns.kdeplot(cdr["logConfirmed"],shade=True,gridsize=200)
sns.kdeplot(cdr["logDeaths"],shade=True,gridsize=200)
sns.kdeplot(cdr["logRecovered"],shade=True,gridsize=200)
plt.legend(["Confirmed Cases","Death Cases","Recovered Cases"])
plt.xlabel("Log(Cases)")
plt.title("KDE Plots of Covid Cases")
# ax = df.plot.kde()
# plt.set(title = "KDE Plots of Covid Cases",xlabel="Log(Cases)")
```

```python
import pandas as pd
newdf = df.groupby('Country_Region')[['Confirmed', 'Deaths', 'Recovered']].sum().astype('int64').sort_values('Confirmed',ascending= False)
newdf.head(10)
```

```python
# Confirmed, Deaths, and Recovered are selected and group by the country_region. The Confirmed, Deaths and Recovered
#cases are converted to integer value, they are added and are sort in descending order by Confirmed Cases.
```

```python
import pandas as pd
cd_remove_zero = newdf[(newdf['Confirmed'] > 0) &  (newdf['Deaths'] > 0)]
cd_remove_zero.tail(5)
```

```python
# Only the rows with confirmed and death cases greater than zero are selected and the countries with least non
#              zero confirmed cases are displayed. 
```

```python
Ques 12. Add a new column named “Fatality Percentage” in the DataFrame from Step 11.

		Fatality Percentage = (Deaths / Confirmed) * 100 

	Show the countries that have more than 2 million Confirmed cases, ranked by ‘Fatality 	Percentage’
```

```python
cd_remove_zero['Fatality_Percentage'] = cd_remove_zero['Deaths'] / cd_remove_zero['Confirmed'] * 100 
cd_remove_zero[cd_remove_zero.Confirmed > 2000000].sort_values('Fatality_Percentage', ascending = False)
```

```python
#Fatality_Percentage column is added and only those countries are displayed where there are more than 2 million 
#     confirmed cases and they are sorted by Fatality_Percentage descending.
```

```python
import pandas as pd
top_deaths = df.groupby('Country_Region')[['Deaths']].sum().sort_values('Deaths',ascending= False).head(5)
top_deaths
```

```python
# At first the deaths are added by country and then top five countries with the highest number of deaths are displayed
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
top_deaths.reset_index(level=0, inplace=True)
plt.bar(top_deaths['Country_Region'], top_deaths['Deaths'],label = "Deaths")
plt.xlabel("Country")
plt.ylabel("Number of Cases")
plt.legend(framealpha=0.5, frameon=True)
plt.show()
```
