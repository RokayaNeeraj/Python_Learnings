#Class variable
A Python class variable is shared by all object instances of a class. They are not defined inside any methods of a class 
because a class variable is shared by instances of a class. As a result, all instances
of the class will be able to access that variable.

```python
#Class variable example
class student():
  category = "Girls" #here, category is the class variable of the class student.
print(student.category)  
student.category = "boys" #changing the class variables value to boys.
print(student.category)
```

Instance Variables

 Instance variables are owned by an instance of a class. The value of an instance variable can be different depending on the instance with which 
 the variable is associated.Instance variables are declared inside a class method.Instance variables are normally defined inside the __init__ method.Init is a method,also called as a constructor which gets called automatically when an instance of the class is created.There are 3 types of methods in Python.

1. Instance methods

2. Class methods

3. static methods



```python
#Instance method example
class student:
  def __init__ (self, name, age): #instantiating two instances of the class
    self.name = name
    self.age = age
first_student = student("ram" , 23)
second_student = student("niraj", 25)
print(first_student.name)
print(first_student.age)
print(second_student.name)
print(second_student.age)
```

```python
# __str__ method
class student:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def __str__(self):
    return '{} age is {}'.format(self.name, self.age)
first_student = student("ram" , 23)
second_student = student("niraj", 25)
print(first_student)
print(second_student)
```

```python
class student:
  def __init__(self, name, jobs):
    self.name = name
    self.jobs = jobs

  def __str__(self):
    return '{} number of jobs is {}'.format(self.name, self.jobs)
  
  def jobs_up(self):
    self.jobs += 1
first_student = student("ram" , 1)
second_student = student("niraj" , 1)
print(first_student)
print(second_student)
first_student.jobs_up()
second_student.jobs_up()
print(first_student)
print(second_student)
```

#Generators
Python generators are a simple way to create iterators. A generator is a function which returns an iterator object. The iterator object can then be iterated over one item at a time.

We can create generator function in Python using the yield keyword. A generator function can contain one or more yield statements, and it may contain return statement as well.

The difference between **return** and **yield** is as follows.

**return** : returns the value and terminates the function.

**yield** : pauses the function saving all its states and continues on from there on subsequent calls. Does not terminate the function.

```python
def gen_example():
  yield "ram"
  yield "shyam"
  yield "hari"
name = gen_example()
print(type(name))
print(next(name))
print(next(name))
print(next(name))
```

```python
print(next(name)) # This line gives error because all the elements are already printed
```

```python
def rev_name(name):
  length = len(name)
  for i in range(length-1, -1, -1):
    yield name[i]
for i in rev_name("Niraj"):
  print(i)
```

#Decorators
A decorator in Python takes in a function, adds some functionality to it and returns it.

Functions are first-class citizens in Python, which means that functions can be treated just like any other regular variables, i.e. functions can be

1) passed as arguments to a another function

2) returned from a function

3) modified

4) assigned to a variable

```python
#Function passed as argument to a another function
def multiply_two(num):
  return num * 2
def divide_two(num):
  return num // 2
def new_func(func , num):
  return func(num)
print(new_func(multiply_two,4)) # multiply_two is passed to the function as an argument
print(new_func(divide_two,4))
```

```python
#Returned from a function
def Bonus():
  def first_staff(salary):
    return salary * 2
  return first_staff
Final_Salary = Bonus()
print(Final_Salary(2000))
```

```python
#Assigning function to a variable
def multiply_two(num):
  return num * 2
new_func = multiply_two
print(multiply_two(6))
print(new_func(6))
```

```python
#Decorator
def decorate_func(func):
  def inner_func():
    print("decorated func")
    return func()
  return inner_func
def normal_func():
  return "anything"
decorated = decorate_func(normal_func)
print(decorated())
```

```python
def divide(func):
  def inner_func(x,y):
    print("Dividing {} by {}".format(x,y))
    if y == 0:
      print("Sorry! Number can't be divided by zero")
      return
    return func(x,y)
  return inner_func
@divide
def div(x,y):
  return x/y
div(2,0)
div(4,2)
```

```python
div(16,0)
div(16,3)
```

#Closure

When we define a function inside another function, the inner function is called a nested function. Nested functions are able to access variables defined in the enclosing (outer) function.

```python
def outer_func():
  output = "This is outer func"
  def inner_func():
    print(output)
  return inner_func
new_func = outer_func()
new_func()
```

```python
del outer_func
```

```python
new_func() 
```

Even if we deleted the original function outer_func(), the output variable defined in the original function was still availble.

This is called Closure.

As seen above, the properties of closure are:

1. There must be a nested function.
2. The nested function should refer to a variable defined in the scope of enclosing function.
3. The enclosing function should return the nested function.

#Computation on NumPy Arrays: Universal Functions

The reason NumPy is so important in the world of Python Data Science is that it provides an easy and flexible interface to optimized computation with arrays of data. What make them so fast are the vectorized operations, which are generally implemented through NumPy's Universal Functions (ufuncs).

```python
# computing reciprocals of an array
import numpy as np
np.random.seed(0)
def compute_reciprocals(values):
  output = np.empty(len(values))
  for i in range(len(values)):
    output[i] = 1 / values[i]
  return output
values = np.random.randint(1,5, size = 5)
compute_reciprocals(values)
```

**Specification of the Merge key**

As we saw, pd.merge() looks for one or more matching column names between the two inputs, and uses this as the key.

However, the column names may not match perfectly in some cases. In such cases pd.merge() provides variety of options for handling this.

**The on keyword**

You can explicitly specify the name of the key column using the on keyword, which takes a column name or a list of column names.

```python
print(df1)
print("---------------")
print(df2)
```

```python
pd.merge(df1, df2,on = 'Student')
```

This option works only if both the left and the right DataFrames have the specified column name.



**The left_on and right_on keywords**

Sometimes you may want to merge two DataFrames on column name/s with different names in the right and the left DataFrame.

For instance, you may have 'name' as student name in one DF and 'student' on the other.

Let's see how we can join on those columns.

```python
# name column will be dropped
x= pd.DataFrame({'name':["ram", "shyam","hari","gita"], 
                 'project':["cs", "ds", "da", "seo"]})
pd.merge(df1, x, left_on = 'Student', right_on = 'name')
```

**The left_index and right_index keywords**

You can also merge on an index rather than a column.

```python
a = df1.set_index('Student') # setting the 'Student' column as explicit index on df1
b = df2.set_index('Student') # setting the 'Student' column as explicit index on df2
print(a)
print('-----------------')
print(b)
```

```python
pd.merge(a, b, left_index=True, right_index=True)
```

The join() method performs a merge that defaults to joining on indices.



```python
a.join(b)
```

**Specifying Set Arithmetic for Joins**

Sometimes there are cases where a value appears in one key column and not on the other.

```python
st1 = pd.DataFrame({"Student":["ram","shyam","hari","gita"],
                    "class": [9 ,10 ,11, 12]})
st2 = pd.DataFrame({"Student":["shiva","ram","shyam","krishna"],
                    "subject": ["mathematics", "statistics", "science", "sociology"]})
print(st1);print('-'*20);print(st2)
```

```python
pd.merge(st1,st2)
```

In the above case, the result contains the intersection of the two sets of inputs, which is known as inner join.



**There are four types of joins.**

1. Inner Join

2. Outer Join

3. Left Join

4. Right Join

```python
pd.merge(st1, st2, how='inner') # explicit inner join (Results Intersection)
```

```python
pd.merge(st1, st2, how = 'outer') # outer join (Results Union)
```

```python
pd.merge(st1, st2, how = 'left') # prints common elements of st1 and st2 and all elements of st1
```

```python
pd.merge(st1, st2, how = 'right') # prints common elements of st1 and st2 and all elements of st2
```

**Overlapping Column Names: The suffixes Keyword**

There may be cases where the two input DataFrames have conflicting column names.

```python
st1 = pd.DataFrame({"student":["a", "b", "c", "d"],
                    "fav_sub": ['mahematics', "computers", "social", "da"]})
st2 = pd.DataFrame({"student":["a","b","c","d"],
                    "fav_sub":["computers","computers","social","social"]})
print(st1);print('-'*20);print(st2)
```

```python
pd.merge(st1, st2, on= "student")
```

If these defaults names are not appropriate, then it is possible to specify a custom suffix using the suffixes keyword.



```python
pd.merge(st1, st2, on = "student", suffixes = ['_one', "_two"])
```

#Aggregate, filter, transform, apply

The GroupBy objects have

1. aggregate()

2. filter()

3. transform()

4. apply()

methods that efficiently implement a variety of useful operations before combining the grouped data.

```python
import numpy as np
import pandas as pd

rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                  'data1': range(6),
                  'data2': rng.randint(0, 10, 6)},
                  columns = ['key', 'data1', 'data2'])
df
```

**1. Aggregation**

We have seen aggregations with sum(), median(), and the like, but aggregate() method allows for even more flexibility.

It can take a function, or a list thereof, and compute all the aggregates at once.

```python
df.groupby('key').aggregate([min, np.median, max])
```

Another useful pattern is that we can pass a dictionary mapping column names to operations to be applied on that column.



```python
df.groupby('key').aggregate({'data1':'min',
                           'data2':'max'})
```

**2. Filtering**

Filtering allows to drop data based on the group properties.

```python
def filter_func(x):
  return x['data2'].std() >4
print(df)
print('-'*20) 
print(df.groupby('key').std())
print('-'*20)
print(df.groupby('key').filter(filter_func))
```

In the above result, key A does not have standard deviation greater than 4, so it is not included (last table).



**3. Transformation**

While aggregation returns a reduced version of the data, transformation can return some transformed version of the full data to recombine.

For such transformation, the output is the same shape as the input.

As example is to center the data by subtracting the group-wise mean.

```python
df.groupby('key').transform(lambda x: x - x.mean())
```

**The apply() method**

The apply() method lets us apply an arbitrary function to the group results. The function takes a DataFrame, and returns either a Pandas object (e.g. DataFrame, or Series) or a scalar; and then the combine operation will be tailored to the type of output returned.

Below is and example of apply() which normalizes the first column by the sum of the second.

```python
def norm_by_data2(x):
  # x is a DataFrame of group values
  x['data1'] /= x['data2'].sum()
  return x

print(df)
print('-'*20) 
print(df.groupby('key').apply(norm_by_data2))
```

**Working with Time Series**

Pandas has extensive set of tools for working with dates, times, and time-indexed data.

Date and time data comes in a few flavors.

1. Time stamps: reference particular moments in time (for example - June 14, 2009 at 6:25 PM).

2. Time intervals and periods: reference a length of time between a particular beginning and end point, for example - the year 2020. Periods usually reference a special case of time intervals in which each interval is of uniform length and does not overlap (e.g. 24 hour-long periods constituting days).

3. Time deltas or durations: reference an exact length of time (e.g. a duration of 26.65 seconds)

**Dates and Times in Python**

Python has a number of available representations of dates, times, deltas, and timestamps.

**Pandas Time Series Data Structures**

The fundamental Pandas data structures for working with time series data are as follows.

1. For time stamps, Pandas provides the Timestamp type. Please note that it is essentially a replacement for Python's native datetime object, however, it is based on the more efficient numpy.datetime64 data type. The associated index structure is DatetimeIndex.

2. For time periods, Pandas provides the Period type. This encodes a fixed frequency interval based on numpy.datetime64 object. The associated index structure is PeriodIndex.

3. For time deltas or duration, Pandas provides the Timedelta type. Timedelta is a more efficient replacement for Python's native datetime.timedelta type, and is based on numpy.timedelta64. The associated index structure is TimedeltaIndex.

The most fundamental of these date/time objects are the Timestamp and DateTimeIndex objects.

While these objects can be invoked directly, it is common to use the pd.to_datetime() function, which can parse a wide variety of formats.

Passing a single date to pd.to_datetime() gives us Timestamp;

Passing a series of dates by default gives us DatetimeIndex.



```python
from datetime import datetime
dates = pd.to_datetime([datetime(2016,2,28), '15th of July, 2020', '2018-Jun-5', '08-06-2017', '20140304'])
dates
```

Any DateTimeIndex can be converted to a PeriodIndex with the to_period() function with the addition of frequency code.

In the following code, we will use 'D' to indicate daily frequency.

```python
dates.to_period('D')
```

A TimedeltaIndex is created, for example, when one date is subtracted from another.

```python
dates - dates[0]
```

**Regular sequences: pd.date_range()**

In order to make the creation of regular date sequences more convenient, Pandas provides a few functions as follows.

1. pd.date_range() for timestamps

2. pd.period_range() for periods

3. pd.timedelta_range() for time deltas.

Just like range() turns a start point, end point, and optional step size into a sequece, pd.date_range() accepts a start date, an end date, and an optional frequency code to create a regular sequence of dates.

By default, the frequency is one day.

```python
pd.date_range('2015-02-03', '2015-02-07')
```

```python
pd.date_range('2014-5-9', periods=7)
```

```python
pd.date_range('2015-9-25', periods=7, freq='H') #hour
```

```python
pd.period_range('2016-7', periods=7, freq='M') #month
```

```python
pd.timedelta_range(0, periods=10, freq='H') #sequence of durations increasing by an hour.
```

**Example: Visualizing Seattle Bicycle Counts**

Let's analyze the data on bicycle counts on Seattle's Fremont Bridge.


```python
!curl -o FremontBridge.csv https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD
```

```python
import pandas as pd

data = pd.read_csv('/content/FremontBridge.csv', index_col='Date', parse_dates=True)
data.head()
```

```python
data.columns = ['Total', 'East', 'West']

data.head()
```

```python
# Summary statistics
data.dropna().describe()
```

**Visualizing the data**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()

data.plot()
plt.ylabel('Hourly Bicycle Count')
```

```python
#Let's resample the data by week
weekly = data.resample('W').sum()
weekly.plot(style=[':', '--', '-'])
plt.ylabel('Weekly bicycle count');
```

The above graph shows some interesting seasonal trends, such as, people bicycle more in the summer than in the winter. Also, the count is going down towards 2020/2021, possibly due to the pandemic.

We can use a rolling mean by employing the pd.rolling_mean() function. Let's do a 30-day rolling mean of our data, making sure to center the window.

```python
daily = data.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style=[':', '--', '-'])
plt.ylabel('mean hourly count');
```

```python
#for smoother version, let's use gaussian window
daily.rolling(50, center=True, win_type='gaussian').sum(std=10).plot(style=[':', '--', '-']);
```

**Digging into data**

While the smoothed data views in the above graph is useful to get an idea of general trend in the data, they hide much of the interesting structure.

For example, we might want to look at the average traffic as a function of the time of the day. We can do that by using the GroupBy functionality.

```python
import numpy as np

by_time = data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks=hourly_ticks, style=[':', '--', '-']);
```

This shows a strongly bimodal distribution, with peaks around 8 AM and 5 PM.

Let's see how things change based on the day of the week.

```python
by_weekday = data.groupby(data.index.dayofweek).mean()
by_weekday.index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
by_weekday.plot(style=[':', '--', '-'])
```

This shows a strong distinction between weekday and weekend totals, with around twice as many average riders crossing the bridge on Monday through Friday than on the weekends.
